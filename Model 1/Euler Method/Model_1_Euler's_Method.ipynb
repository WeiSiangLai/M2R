## Euler's Method

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
epsilon = 0.7
pi = 0.4
t0, tfinal, dt = 0.0, 2.0, 0.1
```

```python
time = np.arange(t0, tfinal+0.001, dt)
P = np.zeros((len(time), 26))
P[0, 10] = 1

for i in range(1, len(time)):
    for j in range(26):
      if not j:
        P[i, 0] = P[i-1, 0] + dt * (epsilon * P[i-1, 1] - pi * P[i-1, 0])
      elif j == 25: # Boundary Conditions
        P[i, 25] = P[i-1, 25] + dt * (pi * P[i-1, 24] - (epsilon + pi) * P[i-1, 25])
      else:
        P[i, j] = P[i-1, j] + dt * (epsilon * P[i-1, j+1] + pi * P[i-1, j-1] - (epsilon + pi) * P[i-1, j])

P /= P.sum(axis=1, keepdims=True)
```

```python
plt.figure(figsize=(8, 5))

for i, col in enumerate(P.T):
    plt.plot(time, col, label=f'$n = {i}$')

plt.title("Evolution of $P_n(t)$ over time (Euler's method)")
plt.xlabel("Time")
plt.ylabel("$P_n(t)$")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.grid()

plt.show()
```
